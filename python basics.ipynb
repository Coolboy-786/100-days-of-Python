```python
def guess_game(a,b):
    jackpot = r.randint(a,b)

    guess = int(input("Enter a number between "+ str(a) + " and " + str(b) + ": "))
    i=0 
    while(jackpot!=guess):
        if (guess < jackpot):
            print("Guess higher")
        else:
            print("Guess lower")
    
        guess = int(input("Enter a number: "))
        i+=1
    
    print( "You guessed it right!!! and it took " + str(i+1) +" attempts to guess it")
```

```python
guess_game(100,200)
```
